# ***INPUT KEYWORD***

```python
user_keyword = input("🔍 Masukkan kata kunci pencarian: ").strip()
tweet_limit = int(input("📊 Masukkan jumlah tweet yang ingin dikumpulkan: ").strip())
search_keyword = f"{user_keyword}"

print("\n✅ Input diterima:")
print(f"🔑 Kata kunci   : {user_keyword}")
print(f"📄 Jumlah tweet : {tweet_limit}")
print(f"📝 Query search : \"{search_keyword}\"")
```

# ***PROSES CRAWLING X***

```python
import subprocess
import time

filename = f"{user_keyword.replace(' ', '_')}.csv"
print("🔄 Sedang melakukan crawling tweet...")

start_time = time.time()
try:
    subprocess.run([
        "npx", "-y", "tweet-harvest@2.6.1",
        "-o", filename,
        "-s", search_keyword,
        "--tab", "LATEST",
        "-l", str(tweet_limit),
        "--token", twitter_auth_token
    ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    end_time = time.time()
    print(f"✅ Crawling selesai dalam {round(end_time - start_time, 2)} detik.")
    print(f"📁 File hasil disimpan sebagai: tweets-data/{filename}")

except Exception as e:
    print(f"❌ Gagal melakukan crawling: {e}")
```

# ***MENAMPILKAN DATAFRAME***

```python
import pandas as pd

# ----------------------------------------
# 📂 Membaca Data dari File CSV
# ----------------------------------------

file_path = f"tweets-data/{filename}"
try:
    df = pd.read_csv(file_path, delimiter=",")
    num_tweets = len(df)
    print(f"\n✅ Jumlah tweet dalam dataframe: {num_tweets}\n")
    print("🔍 5 Tweet pertama:")
    display(df.head())

except Exception as e:
    print(f"❌ Gagal membaca file: {e}")
```

# ***CLEANING***

```python
import re

# ----------------------------------------
# 🔧 Membersihkan Teks
# ----------------------------------------

def cleaning(text):
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = text.strip()
    return text

df_clean = df[['full_text']].copy()
df_clean['cleaned'] = df_clean['full_text'].astype(str).apply(cleaning)

print("\n🔧 Setelah Pembersihan Teks (Cleaning):")
display(df_clean[['full_text', 'cleaned']].head())
```

# ***CASE FOLDING***

```python
# ----------------------------------------
# 🔠 Case Folding (mengubah ke huruf kecil)
# ----------------------------------------

def case_folding(text):
    return text.lower()

df_clean['case_folded'] = df_clean['cleaned'].apply(case_folding)

print("\n🔠 Setelah Case Folding (Semua Huruf Kecil):")
display(df_clean[['cleaned', 'case_folded']].head())
```

# ***TOKENISASI***

```python
# ----------------------------------------
# 🔑 Tokenisasi
# ----------------------------------------

def tokenize(text):
    return text.split()

df_clean['tokenized'] = df_clean['case_folded'].apply(tokenize)

print("\n🔑 Setelah Tokenisasi:")
display(df_clean[['case_folded', 'tokenized']].head())
```

# ***FILTERING & STEMMING***

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

stop_factory = StopWordRemoverFactory()
stopwords = set(stop_factory.get_stop_words())
stem_factory = StemmerFactory()
stemmer = stem_factory.create_stemmer()

def filtering_stemming(tokens):
    filtered = [word for word in tokens if word not in stopwords]
    stemmed = [stemmer.stem(word) for word in filtered]
    return stemmed

df_clean['final_tokens'] = df_clean['tokenized'].apply(filtering_stemming)

print("\n📚 Setelah Filtering dan Stemming:")
display(df_clean[['tokenized', 'final_tokens']].head())
```
